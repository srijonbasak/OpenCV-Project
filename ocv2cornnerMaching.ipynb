```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
chess = cv2.imread('chess.png')
chess = cv2.cvtColor(chess,cv2.COLOR_BGR2RGB)
plt.imshow(chess)
```

```python
chess_gray = cv2.cvtColor(chess,cv2.COLOR_RGB2GRAY)
plt.imshow(chess_gray, 'gray')
```

```python
real_chess = cv2.imread('real_chess.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
real_chess_gray = cv2.cvtColor(real_chess,cv2.COLOR_RGB2GRAY)
plt.imshow(real_chess)
```

```python
plt.imshow(real_chess_gray, 'gray')
```

```python
gray = np.float32(chess_gray)
```

```python
gray
```

```python
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.01)
dst = cv2.dilate(dst,None)
chess[dst>0.01*dst.max()] = [255,0,0]
plt.imshow(chess)
```

```python
gray1 = np.float32(real_chess_gray)
dst = cv2.cornerHarris(src=gray1,blockSize=2,ksize=7,k=1)
dst = cv2.dilate(dst,None)
real_chess[dst>0.1*dst.max()] = [255,0,0]
plt.imshow(real_chess)
```

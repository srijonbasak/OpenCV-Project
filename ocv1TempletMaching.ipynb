```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('cat.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
face = cv2.imread('cat-head.jpg')
face = cv2.cvtColor(face,cv2.COLOR_BGR2RGB)
plt.imshow(face)
```

```python
methods =['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
for x in methods:
    copy_img = img.copy()
    method = eval(x)
    res = cv2.matchTemplate(copy_img,face,method)
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc

    #possition for triengle
    height,width,channel = face.shape
    bottom_right = (top_left[0]+width,top_left[1]+height)
    
    #drawing facr
    cv2.rectangle(copy_img,top_left,bottom_right,(255,55,255),50)
    
    #show image
    plt.subplot(121)
    plt.imshow(res)
    plt.title('result')
    plt.subplot(122)
    plt.imshow(copy_img)
    plt.title('img')
    plt.suptitle(x)
    plt.show()
    print('/n')
```

```python
img1 = cv2.imread('cat.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('dontcopy.png')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
large_image = img1
small_image = img2
```

```python
x_ofset = 0
y_ofset = 0
```

```python
x_end = x_ofset + small_image.shape[1]
y_end = y_ofset + small_image.shape[0]
```

```python
large_image[y_ofset:y_end,x_ofset:x_end] = small_image
```

```python
plt.imshow(large_image)
```
